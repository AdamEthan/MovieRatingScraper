```python
import pandas as pd
import numpy as np
df = pd.read_csv("rotten_tomatoes_top_movies.csv", engine='python')
```

```python
df
```

```python
df=df.drop(columns=df.columns[0])
df=df.drop(columns=["synopsis","consensus","sound_mix","aspect_ratio","writer","view_the_collection","crew","link","release_date_(theaters)","release_date_(streaming)","original_language","production_co"])
```

```python
df['critic_score']= df['critic_score'].fillna(0)
df['people_score']= df['people_score'].fillna(0)
df['people_score']=df['people_score'].astype(np.int64)
df['critic_score']=df['critic_score'].astype(np.int64)
df['total_reviews']=df['total_reviews'].astype(np.int64)
df
```

```python
df.to_csv('movies1.csv', encoding='utf-8')
```
